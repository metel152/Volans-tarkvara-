```python
csv_data = """timestamp,temperature_C,pressure_hPa,gps_altitude_m
2026-03-15 10:00:00,22.0,1013.2,10
2026-03-15 10:01:00,22.1,1013.0,12
2026-03-15 10:02:00,22.3,1012.8,14"""

with open('combined_data.csv', 'w') as f:
    f.write(csv_data)
```

Файл `combined_data.csv` создан. Теперь запускаем код для построения графика.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from CSV file
data = pd.read_csv("combined_data.csv")

# Convert timestamp column to datetime format (optional, but good practice)
data["timestamp"] = pd.to_datetime(data["timestamp"])

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plot temperature on the left Y-axis
ax1.plot(data["gps_altitude_m"], data["temperature_C"],
         color="red", marker="o", label="Temperature (°C)")
ax1.set_xlabel("Altitude (m)")
ax1.set_ylabel("Temperature (°C)", color="red")

# Create secondary Y-axis for pressure
ax2 = ax1.twinx()

# Plot pressure on the right Y-axis
ax2.plot(data["gps_altitude_m"], data["pressure_hPa"],
         color="blue", marker="o", label="Pressure (hPa)")
ax2.set_ylabel("Pressure (hPa)", color="blue")

# Add title and grid
plt.title("Temperature and Pressure vs Altitude")
plt.grid(True)

# Display the plot
plt.show()
```
